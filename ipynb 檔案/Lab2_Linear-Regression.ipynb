### <font color='blue'> Example 2:</font>
### In this problem we will investigate the t-statistic for the null hypothesis $H_0:\beta_1 = 0$ and $H_0:\beta_2 = 0$ in multiple linear regression. To begin, we generate  predictors $x1,\ x2$ and a response $y$ as follows.

#### <font color='green'>  set.seed(1)</font>
#### <font color='green'>  x1=rnorm(100)</font>
#### <font color='green'>  x2=rnorm(100)</font>
#### <font color='green'>  y=5+2*x1+3*x2+rnorm (100)</font>

### <font color='blue'> (a.)</font> Write out the form of the multiple linear model.
### <font color='blue'> (b.)</font> Create a scatterplot displaying the relationship between $x2$ and $y$. Comment on what you observe.
### <font color='blue'> (c.)</font> Perform a multiple linear regression of $y$ onto $x1$ and $x2$. Report the coefficient estimate, the standard error of this coefficient estimate, and the t-statistic and p-value associated with the null hypothesis $H_0:\beta_1 = 0$ and $H_0:\beta_2 = 0$. 
### Comment on these results. (You can use the command by  <font color='red'> lm(y∼x1+x2)</font> .)

### <font color='darkred'> Solution 1:</font>
### <font color='darkred'> (a.)</font>

```r
set.seed(1) #固定亂數種子，讓你的模擬r結果能夠可重複出現
x1=rnorm(100) #產生100個標準常態(N(0,1))的值當成一個解釋變數
x2=rnorm(100)
y=5+2*x1+3*x2+rnorm (100) #反應變數
```

### The form of the multiple linear model is $y=5+2\times x1+3\times x2+\epsilon$, and $\epsilon\sim N(0,1)$.

### <font color='darkred'> (b.)</font>

```r
plot(x2, y)
```

### The relation between $x2$ and $y$ looks linear and with a positive slope.

### <font color='darkred'> (c.)</font>

```r
lm_fit = lm(y~x1+x2) #Perform regression without an interecept
summary(lm_fit) #print the results
```

### According to the summaary result above, we have：
### $\hat{\beta}_0=5.0254$, $\hat{\beta}_1=2.0211$, and $\hat{\beta}_2=2.9465$.
### $se(\hat{\beta}_0)=5.0254$, t-statistic is $47.77$ and the p-value is less than $2\times10^{-16}$.
### $se(\hat{\beta}_1)=2.0211$, t-statistic is $17.31$ and the p-value is less than $2\times10^{-16}$.
### $se(\hat{\beta}_2)=2.9465$, t-statistic is $26.91$ and the p-value is less than $2\times10^{-16}$.
### Under $\alpha=0.05$,  since their p-value are very low we can reject $H_0:\beta_1 = 0$ and $H_0:\beta_2 = 0$ so that $x1$ and $x2$ are highly significant.
